```python
def gettiktok(country,typez):
    api = TikTokApi()

    tiktoks = []
    for vid in api.hashtag(name = [country,typez]).videos(count = 50):
        tiktok = {}
        info = vid.info()
        tiktok['url'] = info['video']['downloadAddr']
        tiktoks.append(tiktok)

    return tiktoks
```
